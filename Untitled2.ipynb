```python
df = df.sort_values(by=['date_time'],ascending=True).reset_index(drop=True)
last_n_hours = [1,2,3,4,5,6]
for n in last_n_hours:
    df[f'last_{n}_hour_traffic'] = df['traffic_volume'].shift(n)
df=df.dropna().reset_index(drop=True)
df.loc[df['is_holiday'] != 'None','is_holiday']=1
df.loc[df['is_holiday'] == 'None','is_holiday']=0
df['is_holiday'] = df['is_holiday'].astype(int)

df['date_time'] = pd.to_datetime(df['date_time'])
df['hour'] = df['date_time'].map(lambda x:int(x.strftime('%H')))
df['month_day'] = df['date_time'].map(lambda x:int(x.strftime('%d')))
df['weekday'] = df['date_time'].map(lambda x:int(x.weekday()+1))
df['month'] = df['date_time'].map(lambda x:int(x.strftime('%m')))
df['year'] = df['date_time'].map(lambda x:int(x.strftime('%Y')))
df.to_csv('traffic_volume_data(1).csv',index=None)

```

```python
from functools import reduce
def unique(list1):
    ans = reduce(lambda re, x: re+[x] if x not in re else re, list1, [])
    print(ans)

n1 = df['weather_type']
n2 = df['weather_description']
unique(n1)
unique(n2)
n1features= ['Rain','Clouds','Clear','Snow','Fog','Mist','Drizzle','Thunderstorm','Squall','Haze']
n2features = ['fewclouds','light rain','Sky is clear','light snow','sky is clear','mist','broken clouds','moderate rain','drizzle','overcast clouds','scattered clouds','haze','proximity thunderstorm','light thunderstorm','thunderstorm with rain','thunderstorm with light rain','thunderstorm with heavy rain','thunderstorm with light drizzle','thunderstorm with drizzle','thunderstorm with heavy drizzle','thunderstorm with light thunderstorm','thunderstorm with heavy thunderstorm','light rain and snow','light intensity shower rain','proximity thunderstorm with drizzle',]

n11 = []
n22 = []
for idx,val in enumerate(n1):
    if val not in n1features:
        n11.append(val)
    else:
        n11.append((n1features.index(val))+1)
    if(n2[idx]) not in n2features:
       n22.append(n2[idx])
    else:
        n22.append((n2features.index(n2[idx]))+1)

df['weather_type'] = n11
df['weather_description'] = n22
```

```python
features = numeric_columns + label_columns
target = ['traffic_volume']
X = df[features]
y = df[target]
```

```python
import pandas as pd
X =pd.get_dummies(X,columns=['weather_description'])

x_scaler = MinMaxScaler()
X = x_scaler.fit_transform(X)
y_scaler = MinMaxScaler()
y = y_scaler.fit_transform(y).flatten()
warnings.filterwarnings('ignore')
```

```python
df['date_time'] = pd.to_datetime(df['date_time'])

df['monthly'] = df['date_time'].dt.to_period('M').dt.to_timestamp()
df['weekly'] = df['date_time'].dt.to_period('W').dt.to_timestamp()
df['daily'] = df['date_time'].dt.to_period('D').dt.to_timestamp()
df['hourly'] = df['date_time'].dt.to_period('H').dt.to_timestamp()







metrics = ['monthly','weekly','daily','hourly']
fig = plt.figure(figsize=(8,4*len(metrics)))
for i,metric in enumerate(metrics):
      ax=fig.add_subplot(len(metrics),1,i+1)
      ax.plot(df.groupby(metric)['traffic_volume'].mean(),'-o')
      ax.set_xlabel(metrics)
      ax.set_ylabel("Mean Traffic")
      ax.set_title(f"Traffic trend by {metric}")
plt.tight_layout()
plt.show()
```

```python
regr = MLPRegressor(random_state=1, max_iter=500).fit(X,y)
new= []
print('predicted_output  =',regr.predict(X))
print('actual_output =',y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.2)
y_pred = regr.predict(X_test)
print('Mean Absolute Error :',mean_absolute_error(y_test,y_pred))
```
